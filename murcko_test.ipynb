```python
from rdkit.Chem.Scaffolds import MurckoScaffold
import rdkit.Chem as Chem
from rdkit.Chem import Draw
from rdkit.Chem import PandasTools
from rdkit.Chem.Draw import IPythonConsole # Enables RDKit IPython integration]
import pandas as pd
```

```python
mol  = Chem.MolFromSmiles('Cc1ncc(C[n+]2csc(CCO[P](O)(=O)O[P](O)([O-])=O)c2C)c(N)n1')
scaffold = MurckoScaffold.GetScaffoldForMol(mol)
scaffold_smiles = MurckoScaffold.MurckoScaffoldSmilesFromSmiles('Cc1ncc(C[n+]2csc(CCO[P](O)(=O)O[P](O)([O-])=O)c2C)c(N)n1')
generic = MurckoScaffold.MakeScaffoldGeneric(MurckoScaffold.GetScaffoldForMol(mol))
Draw.MolsToGridImage([mol, scaffold, generic])
```
