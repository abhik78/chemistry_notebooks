```python
import rdkit.Chem as Chem
from rdkit.Chem import Draw
from rdkit.Chem import PandasTools
from rdkit.Chem.Draw import IPythonConsole # Enables RDKit IPython integration]
import pandas as pd
```

```python
drug_like = pd.read_csv('scaffold_sorted_mcule_drugbank_drugcentarl_chembl.csv')
```

```python
drug_like
```

```python
drug_like = drug_like.dropna()
```

```python
smiles = 'smiles'
count = 'count'
```

```python
PandasTools.AddMoleculeColumnToFrame(drug_like, smilesCol='smiles')
```

```python
drug_like = drug_like[[smiles, count, 'ROMol']]
drug_like.describe()
```
